```python
%matplotlib widget

import os, glob, re
from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt

import xrayutilities as xu

import ipywidgets as iumv gz 0pw
from ipywidgets import interact

import fabio
from silx.io.specfile import SpecFile

from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.widgets import RectangleSelector
```

```python
base_path = r'/msd_data/checkout/bl17-2/Aug2023_Li/NNO_STO'
data_path = os.path.join(base_path, 'energy_scans')
spec_fname = os.path.join(data_path, 'energy_scan')
```

```python
sf = SpecFile(spec_fname)
```

```python
def plot_xas(scan_num):
    scan_data = sf[scan_num-1]
    x_col = 'energy'
    x_data = scan_data.data_column_by_name(x_col)

    y_col = 'i2'
    y_data = scan_data.data_column_by_name(y_col)/scan_data.data_column_by_name('i1')
    

    y = 1/np.log(y_data)
    plt.plot(x_data, y, 'o-', label = scan_num)
    return
```

```python
plt.figure()

plot_xas(1)
plot_xas(2)
plot_xas(3)
plot_xas(4)
```
